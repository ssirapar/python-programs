```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,12,9,16,6,14,18])
y= np.array([4,18,5,18,2,10,8])

plt.plot(x, y)
plt.show()
plt.bar(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 12])

plt.plot(ypoints, marker = '*')
plt.show()
```

```python
#for breaking of line
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, '*:r',ms=15,mec='b')      #mec gives the out colour to the marker  ,  : gives dotted line
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, marker = '*', ms = 15,mec='hotpink',mfc='hotpink')  #mfc gives the colour to marker
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([89,36,85,74,65,96,93,58])
marks = ['chem','clad','it','c program','single','several','workshop','vdc']


plt.pie(y,labels=marks,startangle=180)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([89,36,85,74,65])
mylabels = ['chem','clad','it','c program','single']
myexplode = [0.2,0,0,0,0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```
